### scraping from review page

```python
driver = webdriver.Chrome('C:/Users/Hanna Kim/Documents/데잇걸즈 4기/공공데이터 공모전/chromedriver')
driver.get("https://mydramalist.com/reviews/shows?sort=helpful")
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
driver.close()
```

```python
# a.text-primary : selects Title, Reviewer information
raw = soup.select('a.text-primary')
```

```python
raw
```

in the list, there are three information : title, reviewer, and link to profile<br>
제목, 리뷰작성자, 프로필 링크 정보가 있음<br>
<br>
we need to make a dataframe using them<br>
이걸 기반으로 데이터프레임을 만들어야 함

```python
frame = pd.DataFrame(columns = ['Title', "Reviewer"]) #create a frame dataframe

for i in range(15): # one page has 15 reviews
    list = [raw[3*i].get_text(), raw[3*i+1].get_text()] #use only title and reviewer
    imsi = pd.DataFrame([list],columns = ['Title', 'Reviewer'])
    frame = frame.append(imsi)
```

```python
soup.select("span.score") #the overall score given by each reviewer
```

```python
overall = soup.select("span.score")
imsi = []
for i in overall:
    imsi.append(i.text) #take each score and append to list
frame['overall'] = imsi  #take list and make a column in dataframe
```

```python
soup.select("span.p-l-md.pull-right")  #other scores(for 4 categories) given by each reviewer
```

```python
frame2 = pd.DataFrame(columns = ['Story', "Acting/Cast", "Music", "Rewatch Value"]) 
raw = soup.select("span.p-l-md.pull-right")
for i in range(15):
    list = [raw[4*i].get_text(), raw[4*i+1].get_text(),raw[4*i+2].get_text(),raw[4*i+3].get_text()]
    imsi = pd.DataFrame([list],columns = ['Story', "Acting/Cast", "Music", "Rewatch Value"])
    frame2 = frame2.append(imsi)
```

```python
frame2
```

```python
pd.concat([frame, frame2], axis=1) #concat two frames
```

### userid를 사용하여 url 조작 + soup 구해서 여러 user의 review 스크래핑 가능하도록

```python
def user(userid):
    driver.get(f"https://mydramalist.com/profile/{userid}/reviews") #함수로 입력된 userid의 review page 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    driver.close()
    return soup
```

### using Class

```python
class PerUser():
    def __init__(self, userid):
        self.userid = userid
        
        
    def get_soup(self):
        driver = webdriver.Chrome('C:/Users/Hanna Kim/Documents/데잇걸즈 4기/공공데이터 공모전/chromedriver')
        driver.get(f"https://mydramalist.com/profile/{self.userid}/reviews")
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        time.sleep(5)
        driver.close()
        return soup
    
    
    def frame(self):
        soup = self.get_soup()
        
        title = []
        reviewer = []
        story = []
        acting = []
        music = []
        rewatch = []
        
        tire = soup.select("a.text-primary")
        score = soup.select("span.p-l-md.pull-right")
        
        for i in range(int(len(tire)/3)):
            title.append(tire[3*i].text)
            reviewer.append(tire[3*i+1].text)
            story.append(score[4*i].text)
            acting.append(score[4*i+1].text)
            music.append(score[4*i+2].text)
            rewatch.append(score[4*i+3].text)
            
        overall = [*map(lambda x:x.text,soup.select("span.score"))]
        
        return pd.DataFrame({"title":title, "reviewer":reviewer, "overall":overall,"story":story,"acting":acting, "music":music, "rewatch":rewatch})
```
